```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading file 
data = pd.read_csv(r'C:\Users\harsh kesharwani\Downloads\netflix_titles.csv.zip')
```

```python
#retreving first 5 records of dataset using head function
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
sns.countplot(data.type)
```

```python
#percentage comparision between movie and TV show
Count_movie = len(data[data.type=='Movie'])
Count_tv = len(data[data.type=='TV Show'])

print("Percentage of movies: {:.2f}%".format((Count_movie / (len(data.type))*100)))
print("Percentage of TV Shows: {:.2f}%".format((Count_tv / (len(data.type))*100)))
```

```python
#top 30
data['director'].value_counts().head(30).plot.bar()
```

```python
data['country'].value_counts().head(30).plot.bar()
```

```python
data['release_year']= pd.to_numeric(data['release_year'])

pd.crosstab(data.release_year, data.type).plot(kind = 'bar', figsize=(20,6))
plt.xlabel('Movies Released Year')
plt.ylabel('Frequency')
plt.title('Type of Movies Released in a Year')
plt.show()
```

```python
pd.crosstab(data.rating, data.type).plot(kind='bar', figsize=(20,10))
plt.xlabel('Movies Rating')
plt.ylabel('Frequency')
plt.title('Rating of various contents on Netflix')
plt.show()
```

```python
#20 most frequent genre for movie type for all the years
plt.style.use("ggplot")
plt.figure(figsize=(12,6))
data[data["type"]=="Movie"]["listed_in"].value_counts()[:20].plot(kind="barh",color="red")
plt.title("20 most frequent genre for movie type for all the years",size=18)
```

```python
#20 most frequent genre for TV show type for all the years
plt.style.use("ggplot")
plt.figure(figsize=(11,6))
data[data["type"]=="TV Show"]["listed_in"].value_counts()[:20].plot(kind="barh",color="darkviolet")
plt.title("20 most frequent genre for TV show type for all the years",size=18)
```

```python
#slicing the data specific to India
#contents that have been released only for Indian viewers and not in other countries
data_india= data[data['country']== 'India']

data_india.head()
```

```python
data_india.type.value_counts()
```

```python
data_india.isnull().sum()
```

```python
#dropping the null values
data_india= (data_india.dropna())
```

```python
#containing 707 records with 11 attributes after dropping null values from indian dataset
data_india.shape
```

```python
#top 30 directors in India
data_india.director.value_counts().head(30).plot.bar()
```

```python
#number of netflix contents in India has increased after 2005
plt.figure(figsize=(20,8))
sns.countplot(data_india.release_year)
plt.xticks(rotation=90)
```

```python
#netflix movies are released at a very high speed at compared to TV shows on netflix in India
plt.figure(figsize=(20,8))
sns.countplot(data_india.release_year, hue= data_india.type)
plt.xticks(rotation=90)
```

```python
#Rating for Movies that are released in India
plt.style.use("ggplot")
plt.figure(figsize=(12,6))
data[(data["country"]=="India") & (data["type"]=="Movie")]["rating"].value_counts().plot(kind="bar",color="deeppink")
plt.title("Rating for Movies that are released in India",size=18)
```

```python
#count of movie and TV show based on time duration
plt.figure(figsize=(20,8))
sns.countplot(data_india.duration, hue= data_india.type)
plt.xticks(rotation=90)
```
